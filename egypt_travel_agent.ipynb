## Agent Configuration

````python
from strands import Agent
from strands.models.bedrock import BedrockModel

EGYPT_TRAVEL_SYSTEM_PROMPT = """
You are an expert Egypt travel assistant. Your role is to help travelers plan
their trips to Egypt by providing personalized recommendations for destinations,
hotels, restaurants, activities, and complete trip plans.

You specialize in:
- Helping users discover Egypt's top travel destinations
- Recommending hotels that match their budget and preferences
- Suggesting restaurants based on cuisine preferences and budget
- Finding activities and experiences at each destination
- Planning complete trips within a specified budget

When interacting with users:
- Be friendly, knowledgeable, and enthusiastic about Egypt travel
- Remember user preferences (destination, budget) throughout the conversation
- Proactively use your tools to provide specific, data-driven recommendations
- Break down costs clearly when presenting trip plans
- If a user's budget is too low, suggest alternatives or partial plans
"""

model = BedrockModel(
    model_id="anthropic.claude-sonnet-4-6",
    region_name="us-west-2"
)

agent = Agent(
    model=model,
    tools=[get_destinations, get_destination_details, get_hotels,
           get_restaurants, get_activities, plan_trip],
    system_prompt=EGYPT_TRAVEL_SYSTEM_PROMPT
)
````

## Interactive Demo

The cell below starts an interactive conversation loop with the Egypt Travel Agent. Type your travel questions and the agent will respond using its tools to look up destinations, hotels, restaurants, activities, and trip plans.

The agent maintains session memory, so it remembers your preferences (like destination and budget) throughout the conversation. You don't need to repeat information you've already provided.

Type `quit`, `exit`, or `q` to end the conversation.

### Example Queries

Here are some sample queries you can try with the agent:

- "What destinations are available in Egypt?"
- "Tell me about Luxor"
- "Find hotels in Cairo under $100 per night"
- "What Egyptian restaurants are in Cairo?"
- "What activities can I do in Hurghada for under $50?"
- "Plan a trip to Aswan with a budget of $200 for 2 nights"

````python
print("Egypt Travel Agent - Type 'quit' or 'exit' to end the conversation")
print("=" * 60)

while True:
    user_input = input("\nYou: ")
    if user_input.strip().lower() in ["quit", "exit", "q"]:
        print("\nThank you for using the Egypt Travel Agent. Have a great trip!")
        break
    response = agent(user_input)
    print(f"\nAgent: {response}")
````
