```python
import asyncio
import threading
import uvicorn
import os
import sys

# Ensure the project directory is on the path
project_dir = os.path.dirname(os.path.abspath('__file__'))
if project_dir not in sys.path:
    sys.path.insert(0, project_dir)

from app import app

PORT = 8000

# Use uvicorn Config+Server to avoid nest_asyncio compatibility issues
def start_server():
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)
    config = uvicorn.Config(app, host='0.0.0.0', port=PORT, log_level='info')
    server = uvicorn.Server(config)
    loop.run_until_complete(server.serve())

server_thread = threading.Thread(target=start_server, daemon=True)
server_thread.start()

import time; time.sleep(1)  # Give server a moment to start

print(f'\n✅ Server running on port {PORT}')
print(f'\n🌐 Open the chat UI:')
print(f'   Local:  http://localhost:{PORT}')
print(f'   AWS:    Use your Jupyter proxy URL, e.g.:')
print(f'           https://<your-domain>/proxy/{PORT}/')
print(f'\n   For SageMaker Studio, the URL is typically:')
print(f'           https://<studio-domain>/jupyter/default/proxy/{PORT}/')
```

## Accessing the UI on AWS Jupyter

On AWS-hosted Jupyter environments (SageMaker Studio, SageMaker Notebook Instances, or EC2-hosted JupyterLab), the server binds to `0.0.0.0:8000` but isn't directly accessible from your browser. Use the **Jupyter proxy**:

- **SageMaker Studio**: `https://<your-studio-url>/jupyter/default/proxy/8000/`
- **SageMaker Notebook Instance**: `https://<notebook-url>/proxy/8000/`
- **EC2 JupyterLab**: `http://<ec2-public-ip>:8000` (if port 8000 is open in your security group)

The chat interface will load at the root path (`/`).

```python
# Optional: Test the server is responding
import requests
try:
    resp = requests.get(f'http://localhost:{PORT}/')
    print(f'Status: {resp.status_code} — Server is healthy!')
except Exception as e:
    print(f'Error: {e}')
```
